**Import Data**
---



```python
# Import Pandas
import pandas as pd

# Create a DataFrame out of the data
df_students = pd.read_csv('/content/sample_data.csv')

# Remove rows with missing data (axis=0 refers to rows)
df_students = df_students.dropna(axis=0, how='any')

# Calculate who passed assuming 60 is the passmark
passes = pd.Series(df_students.StudyHours >=60)

# Concatenate the Series to the initial DataFrame (axis=1 refers to )
newDataFrame = pd.concat([df_students, passes.rename('HasPassed')], axis=1)

newDataFrame
```

**Visualizing data with Matplotlib**
---

```python
# Ensuring plots are displayed inline in the Notbook
%matplotlib inline

from matplotlib import pyplot as plt

# Create a bar plot of name vs grade
plt.bar(x=newDataFrame.Name, height=newDataFrame.Grade, color='orange')

# Customise the graph
plt.title('Student Grades')
plt.xlabel('Student')
plt.ylabel('Grade')
plt.grid(color='#95a5a6', linestyle='--', linewidth=1, axis='y', alpha=0.7)
plt.xticks(rotation=90)

# Display the plot
plt.show()
```
