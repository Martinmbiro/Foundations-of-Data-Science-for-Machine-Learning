```python
# Pandas is a more convenient library when working with Data Sets
import pandas as pd
import numpy as np

# Sample student grades below inserted in a list:
data = [50,50,47,97,49,3,53,42,26,74,82,62,37,15,70,27,36,35,48,52,63,64]

# Sample students' study hours below
study_hours = [10.0,11.5,9.0,16.0,9.25,1.0,11.5,9.0,8.5,14.5,15.5,
               13.75,9.0,8.0,15.5,8.0,9.0,6.0,10.0,12.0,12.5,12.0]

# Converting data into NumPy Array of grades
grades = np.array(data)

# Create a 2D array of student data
student_data = np.array([study_hours, grades])

# Creating a DataFrame with three columns
df_students = pd.DataFrame({'Name': ['Dan', 'Joann', 'Pedro', 'Rosie', 'Ethan', 'Vicky', 'Frederic', 'Jimmie',
                                     'Rhonda', 'Giovanni', 'Francesca', 'Rajab', 'Naiyana', 'Kian', 'Jenny',
                                     'Jakeem','Helena','Ismat','Anila','Skye','Daniel','Aisha'],
                            'StudyHours':student_data[0], 'Grade':student_data[1]})
df_students
```

```python
# Retrieving data from a specific index value in a DataFrame using the loc[] method
df_students.loc[5]
#print("\n\n")

# Get rows in range 0-5
df_students.loc[0:5]
```

```python
# Get data in the first five rows using the iloc[] method
# iloc[] returns based on postion, whereas loc[] returns based on index value
df_students.iloc[0:5]

# find the values for the columns in positions 1 and 2 in row 0, like this
df_students.iloc[0,[1,2]]
```

```python
# Using the loc[] method on columns
"""
In the absence of an explicit index column, the rows in our DataFrame are indexed as integer values,
but the columns are identified by name
"""
df_students.loc[0,['Name','StudyHours','Grade']]
```

```python
# find indexed rows based on a filtering expression that references named columns other than the index
df_students[df_students.StudyHours == 1]
#df_students[]
```
