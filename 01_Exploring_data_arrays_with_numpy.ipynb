# **`NumPy`**
> [`NumPy`](https://numpy.org/devdocs/) is a Python library that provides a multidimensional array object, derive objects, an assortment of mathematical tools and functions.

```python
# Suppose a college professor takes a sample of student grades from a class to analyze
data = [50,50,47,97,49,3,53,42,26,74,82,62,37,15,70,27,36,35,48,52,63,64]
print(data)
```

```python
"""
  Converting the above python list into a numpy array to optimize it for data analysis
"""
import numpy as np
grades = np.array(data)
print(grades)
```

```python
"""
The difference between python lists and numpy arrays when both are multiplied by 2
"""

# Multiplying a python listcreates a new list of twice the length with the
# original sequence of list elements repeated
print(f"{type(data)}, * 2: \n{data*2}")

print('---\n')
# Multiplying a NumPy array, on the other hand, performs an element-wise calculation
print(f"{type(grades)}, * 2: \n{grades*2}")
```

```python
# Viewing array shape
grades.shape
```

```python
# Define an array of study hours
study_hours = [10.0,11.5,9.0,16.0,9.25,1.0,11.5,9.0,8.5,14.5,
               15.5,13.75,9.0,8.0,15.5,8.0,9.0,6.0,10.0,12.0,12.5,12.0]

# Create a 2D array (an array of arrays)
student_data = np.array([study_hours, grades])

# display the array
student_data
```

```python
# Viewing the shape of the 2D array
"""
  The student_data array contains two elements,
    each of which is an array containing 22 elements.
"""
student_data.shape
```

```python
# Accessing the first element of the first element
student_data[0][0]
```

```python
# Get the mean value of each sub-array (rounded to 2 decimal places)
mean_hours = round(student_data[0].mean(),2)
mean_grade = round(student_data[1].mean(), 2)

print(f"Average Study Hours: {mean_hours} \nMean Grade: {mean_grade}")
```
